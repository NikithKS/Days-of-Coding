```python
#for p = 0.3

winProb = 0.3
loseProb = 1 - winProb
loss0 = winProb ** 4
loss1 = 4 * (winProb ** 4) * loseProb
loss2 = 10 * (winProb ** 4) * (loseProb ** 2)
duceWin = (20 * (winProb ** 5) * (loseProb) ** 3) / (1 - (2 * winProb * loseProb))

ans = loss0 + loss1 + loss2 + duceWin
print(ans)
print("Percentage= {0:.3g}%".format(ans * 100))
```

```python
xAxis = []
yAxis = []
# for winProb in range(0.3, 0.71, 0.01):
winProb = 0.3
while(winProb <= 0.7):
    xAxis.append(winProb)
    loseProb = 1 - winProb
    loss0 = winProb ** 4
    loss1 = 4 * (winProb ** 4) * loseProb
    loss2 = 10 * (winProb ** 4) * (loseProb ** 2)
    duceWin = (20 * (winProb ** 5) * (loseProb) ** 3) / (1 - (2 * winProb * loseProb))

    ans = loss0 + loss1 + loss2 + duceWin
    yAxis.append(ans)
    winProb += 0.01
```

```python
import matplotlib.pyplot as mplot
mplot.plot(xAxis, yAxis)
mplot.xlabel('Probability of win')
mplot.ylabel('Percentage of winning matches')
mplot.show
```
